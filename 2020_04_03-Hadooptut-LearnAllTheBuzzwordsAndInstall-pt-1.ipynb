## Usando hadoop
0. Descargar datos a usar.
 * En este caso se usan archivos .DATA (Que son tab delimited values), también hay archivos .ITEM, que tienen diferentes formatos, en este ejemplo el .item usa | como separador

1. Una vez que entras al explorador de internet haces click en LAUNCH DASHBOARD y usas ambari con user: maria_dev y password: maria_dev (son los defaults) y corres el dashboard.

3. **Usar Query en las bases de datos**: te vas a la sección de query y usas la Worksheet:
  *  3.1 Puedes usar simples SQL commands, eg: (get a list of most popular to least popular movie (popular = # of times rated))
    ```
    SELECT movie_id, count(movie_id) as ratingCount #the count of each movie is something we sort by
    FROM ratings
    GROUP BY movie_id #group rogheter all of the rating for a given movie_id and count them all up
    ORDER BY ratingCount
    DESC
    ```
  This query basically counts how many times each movie was rated in our data and sort it in reverse order by number of times it showed up

4. **Visualizar el resultado de los datos**: Usas en la derecha el ícono de una gráfica y puedes pasar las columnas que quieres comparar en scatterplot (Positional section)





#Parte 5. Hadoop Overview and History


Pilars of hadoop: 
* Open source: Free
* Software plataform:
  - It is a bunch of software that runs on a cluster of computers. 
  - Hadoop is designed to run in an entire cluster of PC's 
  - it leverages the power of multiple PC's
* Distributed storage 
  - You can keep adding more and more Computers to your cluster and getting more storage/Power. 
  - Hadoop gives you a way of viewing all of the data distributed along all the hard-drives as a single file system 
  - Hadoop allows for error management because it keeps backups of its data on other places in the cluster (in case a Computer/Server in you net malfunctions)
* Distributed processing
  - It can distribute the processing of data among Computers.
  - Hadoop provides the means of allowing a cluster of computers to work on the same problem in parallel.

* Very large datasets/Computer clusters: 
  - Designed to manage large amounts of data with large ammounds of computers

## The core hadoop ecosystem

1. **HDFS**: Hadoop Distributed File System. This could be analog to the *distributed data storage *
  - This system is what allow us to distribute the storage of big data across our cluster of computers.
    + It basically makes all our hard drives in our cluster look like a giant filesystem. 
    + it also mantains backups of data.
    
  - 1.1 **MESOS**:
    + Mesos is an alternative to YARN

2. **YARN**: Yet Another Resource Negotiator: (Runs on top of HDFS)
  - This is where the data processing part of Hadoop comes into play. It is basically the system that manages resources in the data cluster.

3. **MapReduce** (Runs on top of YARN)
  - This is basically a programming model that allows the processing of data across an entire cluster.
    + It consists of mappers and reducers.
      * Mappers transform data in parallel across the entire cluster.
      * Reducers aggregate the data togheter.

  - 3.1 **Pig**, sits on top of MapReduce.
    + This basically allows the user to write scripts in a very similar language to SQL insted of java or python.
  
  - 3.2 **Hive**, sits on top of MapReduce, or TEZ.
    + Hive is a way of actually taking SQL queries and making this data look like a SQL database, it basically allows for everything to work like a relationalDatabase (SQL) even tough it is not really.

4. **Apache Ambari**, this basically sits on top of everything.
  + Ambari gives you a view of your cluster and lets you visualize:
    * what's running on your cluster.
    * What systems we are using
    * Resources used.
  + Also allows you to execute Hive/pig queries, or upload files, etc..

5. **SPARK**, sits on top of YARN (or MESOS)
  + "if you need to quickly, efficiently and reliable process data on your cluster, use SPARK"
    - It can handle SQL queries that can do machine learning across an entire cluster of information.
    - It can handle streaming data in real time and all sorts of other cool stuff.

6. **TEZ**, similar to spark but it uses directed acyclic graph.
  - Tez has a leg up on what map reduce does because it can produce more optimal plans for actually executing queries. Tez is usually used in conjuction with Hive to accelerate it.

7. **HBASE**, its basically a NoSQL database.
  - HBase is a way of exposing the data on the cluster to transactional platforms.
  - (NoSQL): It is a columnar data store, basically a really fast database designed for very fast transaction rates.
    + It is usually useful to hit from a website and do all types of transactions.
  - It basically exposes data stored in the cluster, maybe the data was transformed and this provided a very fast way of exposing those results to other systems.

8. **Apache Storm**: Basically a way to process streaming data (real-time processing) 

9. **OOZIE**: A way of scheduling jobs on your cluster.
  - If you have a task that happens in your cluster and it involves many different steps and systems, oozie can schedule this togheter into jobs that can be run.
    + Basically when you have more comllicated operations that require loading data into hive and then integrating that with pig and querying it with spark and then transforming the results into HBase, etc... Oozie can manage that all for you and make sure it all runs reliably.

10. **Zookeeper** Basically a technology for coordinating all your clusters.
  - It keeps track of which nodes are up/down, of shared states across the cluster, the current master node, etc...
  - many applications rely on zookeeper to mantain realiable and consistence performance in a cluster even when a node goes down.

11. **Data ingestion systems**
  * 11.1 **Sqoop**
    - This can tie your Hadoop database into a relational database.
    - Sqoop is basically a connector between hadoop and your legacy databases.
  * 11.2 **FLUME**
    - Is a way of actually transporting Web Logs at a very large scale and very reliably to your cluster
      - For example you have a fleet of web servers Flume can actually listen to the web logs coming in from those web servers in real time and publish them into your cluster in real time for processing by something like Storm or Spark streaming
  * 11.3 **Kafka**
    - Basically collects data from any sort from a cluster of PCs from a cluster of web server or anything and broadcast that into the hadoop cluster
